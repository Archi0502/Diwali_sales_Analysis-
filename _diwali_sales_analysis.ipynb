# Exploratory Data Analysis


# Gender

```python
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gen = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by ='Amount', ascending = False)

yx=sns.barplot(x='Gender', y='Amount',data= gen)

for bars in yx.containers:
    yx.bar_label(bars)
```

#as we can see most of the buyers are females and purchasing power of females is also more than  males

# Age


```python
ax =sns.countplot(data=df,x='Age Group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by ='Amount', ascending=False)

z=sns.barplot(x='Age Group', y='Amount',data=age)

 
```

# Mrital_status

```python
ax = sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(6,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
marital=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})

sns.barplot(x='Marital_Status',y='Amount',data=marital,hue='Gender')
```


as we can see married people has spent more than single ones

# state

```python
##top 10 state of more purchasing power

state= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,9)})
sns.barplot(x='State',y='Orders',data=state)
 
```

```python
# total amount spent through top 10 states

Sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(data=Sales_state,x='State',y='Amount' )
```

Maximum orders have been placed from uttar pradesh then maharastra and then karnataka show here is top 10 states and we can also see from above analysis that maximum amount has also been spent through these three states.

# occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df, x= 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
 cons_jobs=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(x='Occupation',y='Amount',data=cons_jobs)
```

 as we can see from above Most of the people are from IT sector and most amount spent by IT sector.

# product category

```python
sns.set(rc={'figure.figsize':(25,10)})
ax=sns.countplot(data=df,x='Product_Category')


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(10,4)})
sns.barplot(x='Product_Category',y='Amount',data=sales_product)
```

As we can see from above that people spent more in food and clothing & Apparel and most sold products are from food and clothing then electronics.
